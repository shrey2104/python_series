```python
# Logic Box
# Pattern Generator and Number Analyzer

print("==============================================")
print(" Welcome to Pattern Generator and Number Analyzer ")
print("==============================================")

while True:

    print("\nMenu")
    print("1. Generate a pattern")
    print("2. Analyze a range of numbers")
    print("3. Exit")

    choice = input("Enter your choice (1-3): ")

    # Pattern Generator
    if choice == '1':

        rows = int(input("Enter the number of rows: "))

        # Input validation
        if rows <= 0:
            print("Invalid row count!")
            print("Pattern generation stopped.")
            continue

        print("\nRight-angled Triangle Pattern:")

        # Nested loop
        for i in range(1, rows + 1):
            for j in range(i):
                print("*", end=" ")
            print()

    # Number Analyzer
    elif choice == '2':

        start = int(input("Enter starting number: "))
        end = int(input("Enter ending number: "))

        if end < start:
            print("Error! Ending number must be greater than or equal to starting number.")
            continue

        total = 0

        print("\nOdd and Even Analysis:")

        for num in range(start, end + 1):

            if num == 0:
                print("0 is neither odd nor even")
            elif num % 2 == 0:
                print(num, "is even")
            else:
                print(num, "is odd")

            # Add every number
            total += num

        print("Sum of all numbers =", total)

    # Exit
    elif choice == '3':
        print("Thank you for using Logic Box!")
        print("Program exited successfully.")
        break

    else:
        print("Invalid choice! Please enter 1, 2, or 3.")
```
