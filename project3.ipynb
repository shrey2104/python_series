```python

# STUDENT DATA ORGANIZER
# Collection Manipulator Project


students = []          
subjects_set = set()   

print("=" * 50)
print("WELCOME TO STUDENT DATA ORGANIZER")
print("=" * 50)

while True:

    print("\nMENU")
    print("1. Add Student")
    print("2. Display All Students")
    print("3. Update Student")
    print("4. Delete Student")
    print("5. Display Subjects Offered")
    print("6. Exit")

    choice = input("Enter your choice (1-6): ")

    # ADD STUDENT
    
    if choice == "1":

        name = input("Enter Name: ")

        age = int(input("Enter Age: "))     

        grade = input("Enter Grade: ")

        student_id = input("Enter Student ID: ")

        dob = input("Enter Date of Birth (DD/MM/YYYY): ")

        subjects = input("Enter Subjects (comma separated): ").split(",")

        subjects = [subject.strip() for subject in subjects]

        student_tuple = (student_id, dob)

        student = {
            "info": student_tuple,
            "name": name,
            "age": age,
            "grade": grade,
            "subjects": subjects
        }

        students.append(student)

        subjects_set.update(subjects)

        print("\nStudent Added Successfully!")

    # DISPLAY STUDENTS
    
    elif choice == "2":

        if len(students) == 0:
            print("\nNo student records found.")

        else:

            print("\n===== STUDENT RECORDS =====")

            for student in students:

               sid, dob = student["info"]

                print(f"\nStudent ID : {sid}")

                # .format()
                print("Name       : {}".format(student["name"]))

                # % formatting
                print("Age        : %d" % student["age"])

                print(f"Grade      : {student['grade']}")

                print("DOB        :", dob)

                print("Subjects   :", ", ".join(student["subjects"]))

                print("-" * 40)

    # UPDATE STUDENT
    
    elif choice == "3":

        sid = input("Enter Student ID to Update: ")

        found = False

        for student in students:

            if student["info"][0] == sid:

                found = True

                print("\nWhat do you want to update?")
                print("1. Age")
                print("2. Grade")
                print("3. Subjects")

                option = input("Enter Choice: ")

                if option == "1":

                    student["age"] = int(input("Enter New Age: "))

                elif option == "2":

                    student["grade"] = input("Enter New Grade: ")

                elif option == "3":

                    new_subjects = input("Enter Subjects: ").split(",")

                    new_subjects = [subject.strip() for subject in new_subjects]

                    student["subjects"] = new_subjects

                    subjects_set.clear()

                    for s in students:
                        subjects_set.update(s["subjects"])

                else:
                    print("Invalid Choice")

                print("Student Updated Successfully!")
                break

        if not found:
            print("Student ID Not Found!")

    # DELETE STUDENT
    
    elif choice == "4":

        sid = input("Enter Student ID to Delete: ")

        found = False

        for i in range(len(students)):

            if students[i]["info"][0] == sid:

                del students[i]      # del keyword

                found = True

                subjects_set.clear()

                for s in students:
                    subjects_set.update(s["subjects"])

                print("Student Deleted Successfully!")

                break

        if not found:
            print("Student ID Not Found!")

    # DISPLAY SUBJECT

    elif choice == "5":

        if len(subjects_set) == 0:

            print("No Subjects Available.")

        else:

            print("\nUnique Subjects Offered:")

            for subject in sorted(subjects_set):
                print(subject)
# EXIT

    elif choice == "6":

        print("\nThank You for Using Student Data Organizer.")
        print("Have a Nice Day!")
        break

    else:

        print("Invalid Choice! Please Enter Between 1-6.")
```
