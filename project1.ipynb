```python
print("welcome to interactive personal data collector")
print("hello")
print("this program collects your personal information")
print("it demonstrates the use of python fundamentals such as")
print("print(), input(), variable, data types")
print ("operators,type casting, type(), and id()")

# your personal details
name = "Aanal"
age = 25
height = 5.2  # height in feet
favourite_number = 14

# Displaying the collected information
print("\n---personal information summary---")
print(f"Name: {name}")
print(f"Age: {age} years")
print(f"height: {height} feet ({height * 30.48:.2f} cm)")  # coverting feet to cm
print(f"Favourite Number: {favourite_number}")

# collect user information
name = input("enter your name: ")       # string
age = int(input("enter your age: "))    # integer
height = float(input("enter your height in meters: "))   # float
is_student = input("are you a student? (yes/no): ").lower() == "yes"  # boolean


# Arithmetic operations
current_year = 2026                                 # substraction
birth_year = current_year - age                     # multiplication
double_age = age * 2                                # division
half_height = height / 2                            # addition 

# construct user-friendly output using formatted strings
print("user information summy")
print(f"Hello {name}, nice to meet you")
print(f"your age {age} years old, so you were born in {birth_year}.")
print(f"Next year, your will be {age}: years old.")
print(f"twice your age is {double_age}.")
print(f"half of your height id {half_height:2f} meters.")
print(f"student status: {'yes' if is_student else 'no'}")

name = "Aanal"
age = 25
height = "5'o"
favourite_number = 14

print("----variable summary----")

print("name:")
print("value =", name)
print("data type =", type(name))
print("memory address =", id(name))

print("age:")
print("value =", age)
print("data type =", type(age))
print("memory address =", id(age))

print("height:")
print("value =", height)
print("data type =", type(height))
print("memory address =", id(height))

print("favourite_number:")
print("value =", favourite_number)
print("data type=", type(favourite_number))
print("memory address =", id(favourite_number))

print("thank you for using the personal data collector!")
print("we hope you enjoyed learning about python variables, data types, and memory addresses.")
print("keep practicing and exploring python to build exciting projects and improve your programming skills.")
print("happy coding!")
```
